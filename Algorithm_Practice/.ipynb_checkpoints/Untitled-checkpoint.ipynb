```python
import numpy as np
dataset = np.random.randint(100,size=1000)
```

```python
import pandas as pd
market_capital = pd.DataFrame(dataset, columns=['market_capital'])
```

```python
market_capital.reset_index(inplace=True)
print(market_capital.head())
```

```python
market_capital.plot(kind='scatter', x='index', y='market_capital', alpha=0.5)
```
