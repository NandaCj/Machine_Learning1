```python
import pandas as pd
import os
file_name = "Stock_AnalysisPrice_CHG.csv"
CSV_FILE = os.path.join(os.path.dirname(os.getcwd()), file_name)
```

```python
PSheet = pd.DataFrame.from_csv(CSV_FILE)
```

```python
PSheet.info()
```

```python
PSheet1 = PSheet.dropna(subset=['Mar_16_17_Stock_Price_PChg'])
PSheet1.info()
```

```python
List = list(PSheet1.sort_values(by='Mar_16_17_Net_Worth_PChg', ascending=False)['Mar_16_17_Net_Worth_PChg'][:20].index)
List += list(PSheet1.sort_values(by='Mar_16_17_Net_Worth_PChg', ascending=True)['Mar_16_17_Net_Worth_PChg'][:20].index)
List += list(PSheet1.sort_values(by='Mar_16_17_Stock_Price_PChg', ascending=False)['Mar_16_17_Stock_Price_PChg'][:20].index)
List += list(PSheet1.sort_values(by='Mar_16_17_Stock_Price_PChg', ascending=True)['Mar_16_17_Stock_Price_PChg'][:20].index)
print(len(List))
```

```python
PSheet1Corr = PSheet1.filter(regex='Mar_16_17.*').corr()
PSheet1Corr['Mar_16_17_Stock_Price_PChg'].sort_values(ascending=True) 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
PSheet1.drop(List).plot(kind='scatter', x='Mar_16_17_Net_Worth_PChg', y='Mar_16_17_Stock_Price_PChg')
```
